```python
# imported matplotlib for visualizations
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Exploring visualizations on districts dataset
plt.figure(figsize=(16, 10))

sns.countplot(y="state",data=districts,order=districts.state.value_counts().index,palette="pastel",linewidth=3)
plt.title("State Distribution",size=18)

sns.despine()
plt.show()
```

```python
#Exploring visualizations on products dataset
plt.figure(figsize = (16,8))

sns.countplot(data=products, y="Provider/Company Name", order=products['Provider/Company Name'].value_counts().index[:10], palette='pastel')
plt.title('Top 10 Provider Platforms with the Most Product in 2020', size=18)
sns.despine()
plt.show()
```
